```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.text_splitter import RecursiveCharacterTextSplitter
import os

def initialize_llm():
  llm=ChatGroq(
    temperature=0,
    groq_api_key= "your_groq_api_key_here",
    model_name= "llama-3.3-70b-versatile"
)
  return llm

def create_vector_db():
  loader=DirectoryLoader("/content/data/", glob= '*.pdf', loader_cls= PyPDFLoader)
  documents=loader.load()
  text_splitter=RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
  texts=text_splitter.split_documents(documents)
  embeddings=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
  vector_db=Chroma.from_documents(texts, embeddings, persist_directory='./chroma_db')
  vector_db.persist()

  print("ChromaDB created and data saved")

  return vector_db


def setup_qa_chain(vector_db, llm):
  retriever=vector_db.as_retriever()
  prompt_templates= """ You are a compassionate mental health chatbot. Respond throughfully to the following question:
  {context}
  User: {question}
  Chatbot: """
  PROMPT=PromptTemplate(template=prompt_templates,input_variables=['context','question'])
  qa_chain=RetrievalQA.from_chain_type(
      llm=llm,
      chain_type="stuff",
      retriever=retriever,
      chain_type_kwargs= {"prompt": PROMPT}

  )
  return qa_chain

def main():
  print("initializing chatbot...")
  llm=initialize_llm()

  db_path="/content/chroma_db"

  if not os.path.exists(db_path):
    vector_db= create_vector_db()
  else:
    embeddings=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
    vector_db=Chroma(persist_directory=db_path,embedding_function=embeddings)

  qa_chain=setup_qa_chain(vector_db, llm)

  while True:
    query=input("\nHuman: ")
    if query.lower() == "exit":
      print("Chatbot: Take care of yourself, Goodbye!")
      break

    response= qa_chain.run(query)
    print(f"chatbot: {response}")

if __name__ == "__main__":
  main()
```

```python
import os
import traceback

from langchain_groq import ChatGroq
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.document_loaders import PyPDFLoader, DirectoryLoader
from langchain_community.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.text_splitter import RecursiveCharacterTextSplitter
import gradio as gr


def initialize_llm():
    api_key = os.getenv("your_groq_api_key_here") or "your_groq_api_key_here"
    return ChatGroq(
        temperature=0,
        groq_api_key=api_key,
        model_name="llama-3.3-70b-versatile"
    )


DB_PATH = "/content/chroma_db"
DATA_DIR = "/content/data"

def build_or_load_vector_db():
    os.makedirs(DATA_DIR, exist_ok=True)

    if not os.path.exists(DB_PATH):
        # Build fresh DB
        loader = DirectoryLoader(DATA_DIR, glob="*.pdf", loader_cls=PyPDFLoader)
        docs = loader.load()
        if not docs:
            print("⚠️ No PDFs found. Retrieval will be disabled (chat will still work).")
            return None

        splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
        chunks = splitter.split_documents(docs)

        embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
        vectordb = Chroma.from_documents(chunks, embedding=embeddings, persist_directory=DB_PATH)
        vectordb.persist()
        print("✅ ChromaDB created.")
        return vectordb
    else:

        embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
        vectordb = Chroma(persist_directory=DB_PATH, embedding_function=embeddings)
        print("✅ ChromaDB loaded.")
        return vectordb


def make_qa_or_direct(vectordb, llm):
    if vectordb is None:

        def direct_llm(query: str):

            msg = [{"role": "user", "content": query}]
            return llm.invoke(msg).content
        return ("direct", direct_llm)


    retriever = vectordb.as_retriever(search_kwargs={"k": 4})

    prompt_tmpl = """You are a compassionate mental health assistant. Be supportive, non-judgmental, and clear.
Use the context if relevant; if it's not relevant, answer from general knowledge and best practices.

Context:
{context}

User: {question}
Assistant:"""

    PROMPT = PromptTemplate(
        template=prompt_tmpl,
        input_variables=["context", "question"]
    )

    qa = RetrievalQA.from_chain_type(
        llm=llm,
        chain_type="stuff",
        retriever=retriever,
        chain_type_kwargs={"prompt": PROMPT},
        return_source_documents=False
    )
    return ("retrieval", qa)


print("🔧 Initializing...")
llm = initialize_llm()
vectordb = build_or_load_vector_db()
mode, qa_or_direct = make_qa_or_direct(vectordb, llm)
print(f"🚀 Mode: {mode.upper()}")


def chatbot_fn(message, history):
    if not message or not str(message).strip():
        return "Please type a message."

    try:
        if mode == "direct":
            return qa_or_direct(message)


        out = qa_or_direct.invoke({"query": message})
        if isinstance(out, str):
            return out

        return out.get("result") or out.get("output_text") or out.get("answer") or str(out)

    except Exception as e:

        print("----- Exception in chatbot_fn -----")
        traceback.print_exc()
        print("-----------------------------------")
        return f"Sorry, something went wrong: {e}"


with gr.Blocks(theme="gradio/huggingface") as app:
    gr.Markdown("# 🧠 ChatBot for Mental Health")
    gr.Markdown(
        "A compassionate chatbot designed to assist with mental well-being.\n\n"
        "⚠️ **Disclaimer:** For serious concerns or emergencies, contact a licensed professional or local helpline."
    )
    gr.ChatInterface(
        fn=chatbot_fn,
        title="Hello! How can I help you today?"
    )
    gr.Markdown(
        "Thank you! Goodbye."
    )


app.launch(show_error=True, debug=True)
```
